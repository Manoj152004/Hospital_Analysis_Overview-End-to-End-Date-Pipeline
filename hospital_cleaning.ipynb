```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\System -02\Downloads\Clean_02_hospital_patients.csv")
df.drop_duplicates()
df.columns.str.lower
df.head(2)
```

```python
df['blood_group'] = df['blood_group'].fillna('Unchecked')
df['doctor_name'] = df['doctor_name'].fillna(df['doctor_name'].mode()[0])
df['insurance_type'] = df['insurance_type'].fillna('No')
```

```python
for i in ['admission_date','discharge_date']:
    df[i] = pd.to_datetime(df[i],dayfirst= True,errors='coerce')
```

```python
df['city'] = df['city'].str.title()
df['readmitted_within_30days'] = df['readmitted_within_30days'].str.title()
```
